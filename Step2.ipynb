## Bước 2: Phân tích dữ liệu (EDA)

```python
import os
import pandas as pd
def load_housing_data():
    csv_path = os.path.join("D:\\NCKH\\housing\\housing.csv")
    return pd.read_csv(csv_path)
```

```python
housing = load_housing_data()
```

#### Minh họa các đại lượng bằng biểu đồ

```python
import matplotlib.pyplot as plt
housing.hist(bins=50, figsize=(20,15))
plt.show()
```

Nhìn vào biểu đồ, ta có:
* longtitude và latitude cho ta 2 khu vực có đông dữ liệu => khu vực này có đông dân cư
* housing_median_age và median_house_value có số dữ liệu ở bar cuối rất cao => dữ liệu còn có thể to hơn nhưng đã bị cắt bằng dữ liệu max trong bảng

#### Minh họa mối tương quan giữa các đại lượng 

```python
from pandas.plotting import scatter_matrix
attributes = ["median_house_value", "median_income", "total_rooms",
              "housing_median_age"]
scatter_matrix(housing[attributes], figsize=(12, 8))
```

Lấy ví dụ mối liên hệ giữa median_income và median_house_value
* Ta có thể dễ dàng rút ra được là thu nhập càng cao thì giá trị nhà càng cao và ngược lại

#### Minh họa bằng biểu đồ chấm

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha = 0.1)
```

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
             s=housing["population"]/100, label="population", figsize=(10,7),
             c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=True,
)
plt.legend()
```
